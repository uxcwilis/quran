```python
from IPython.core.inputtransformer2 import tokenize
from keras.preprocessing.text import Tokenizer
#from keras.preprocessing.sequence import pad_sequences
from keras.utils import pad_sequences
from nltk.corpus import stopwords
import re
import nltk
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import pandas as pd

nltk.download('stopwords')

# Create a stemmer object for Indonesian language
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Define a function for punctuation removal
def remove_punctuation(DF_train):
    return re.sub(r'[^\w\s]', '', DF_train)

# Define a function for case folding
def case_folding(DF_train):
    return DF_train.lower()

# Define a function for tokenization, stopword removal, and stemming
# Note: You should create or find an appropriate list of Indonesian stopwords
english_stopwords = set(stopwords.words('english')) # Example stopwords, please update accordingly

def preprocess_text(DF_train):
    # Remove punctuation and case folding
    DF_train = remove_punctuation(case_folding(DF_train))

    # Tokenize and remove stopwords
    tokens = DF_train.split()
    tokens = [token for token in tokens if token not in english_stopwords]

    # Stemming
    stemmed_text = stemmer.stem(' '.join(tokens))
    stemmed_text = ' '.join(tokens)


    return stemmed_text

# Preproses data teks
xbaru_preprocessed = xbaru['Terjemahan'].apply(preprocess_text)

# Proses tokenisasi, konversi teks menjadi urutan, dan padding
MAX_NB_WORDS = 50000

MAX_SEQUENCE_LENGTH = 250
EMBEDDING_DIM = 100
tokenizer = Tokenizer(num_words=MAX_NB_WORDS, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)
tokenizer.fit_on_texts(xbaru_preprocessed.values)
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))

X = tokenizer.texts_to_sequences(xbaru_preprocessed.values)
X = pad_sequences(X, maxlen=MAX_SEQUENCE_LENGTH)
print('Shape of data tensor:', X.shape)
```

```python
xval_preprocessed = DF_val['Terjemahan'].apply(preprocess_text)
Y_val = pd.get_dummies(DF_val['c4']).values

Xval = tokenizer.texts_to_sequences(xval_preprocessed.values)
Xval = pad_sequences(Xval, maxlen=MAX_SEQUENCE_LENGTH)
print('Shape of data tensor:', Xval.shape)
```

```python
Y = pd.get_dummies(xbaru['c4']).values
print('Shape of label tensor:', Y.shape)
```

```python
#X_train, X_val, Y_train, Y_val = train_test_split(X,Y, test_size = 0.10, random_state = 42)

X_train= X
X_val = Xval
Y_train= Y


print(X_train.shape,Y_train.shape)
print(X_val.shape,Y_val.shape)
```

```python
X_train
```

```python
from keras.models import Sequential
from keras.layers import Embedding, SpatialDropout1D, LSTM, Dense
from keras.callbacks import ModelCheckpoint

model = Sequential()
model.add(Embedding(MAX_NB_WORDS, EMBEDDING_DIM, input_length=X.shape[1]))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(286, activation='relu'))
model.add(Dense(2, activation='sigmoid'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Tentukan callback untuk menyimpan model pada epoch terbaik
checkpoint_path = 'best_model.c4'
checkpoint = ModelCheckpoint(checkpoint_path, monitor='val_accuracy', save_best_only=True, mode='max', verbose=1)

epochs = 50
batch_size = 64

#model.fit(X, Y, epochs=epochs, batch_size=batch_size , validation_split=0.1 )
model.fit(X_train, Y_train, batch_size = batch_size, validation_data = (X_val, Y_val), epochs = epochs, callbacks=[checkpoint])
```

```python
from keras.models import load_model

# Memuat model terbaik yang telah disimpan
best_model = load_model('best_model.c4')
```
